## Method 1:to generate list of names that are similar. This method has been adopted from 

https://community.dynamics.com/crm/b/mannygrewaldynamicscrm/archive/2017/08/26/dedupe-duplicates-using-fuzzy-proximity-search

## which uses fuzzy proximity search to find the names that are similar. Output is a csv file with each row having names of people which are similar. This was used to understand fuzzy search that has been explored later in method 2

```python
def pros(df1):
    FULL_MATCHING_THRESHOLD = 80
    PARTIAL_MATCHING_THRESHOLD = 100
    SORT_MATCHING_THRESHOLD = 100
    TOKEN_MATCHING_THRESHOLD = 100
    MAX_MATCHES=4
    current_db_dataframe=df1
    
    def find_matches(matchThis):
        rows = current_db_dataframe['full'].values.tolist();
        rows.remove(matchThis)
        matches= process.extractBests(matchThis,rows,scorer=fuzz.ratio,score_cutoff=FULL_MATCHING_THRESHOLD,limit=MAX_MATCHES)
        if len(matches)==0:
            matches= process.extractBests(matchThis,rows,scorer=fuzz.partial_ratio,score_cutoff=PARTIAL_MATCHING_THRESHOLD,limit=MAX_MATCHES);
            if len(matches)==0:
                matches= process.extractBests(matchThis,rows,scorer=fuzz.token_set_ratio,score_cutoff=TOKEN_MATCHING_THRESHOLD,limit=MAX_MATCHES);
                if len(matches)==0:
                    matches= process.extractBests(matchThis,rows,scorer=fuzz.token_sort_ratio,score_cutoff=SORT_MATCHING_THRESHOLD,limit=MAX_MATCHES);
        return matches[0][0] if len(matches)>0 else None


    fn_find_matches = lambda x: find_matches(x)
    current_db_dataframe['Duplicate']=current_db_dataframe.applymap(fn_find_matches)
    current_db_dataframe.to_csv("using crm crafted method.csv", index=False) #saving results to external file
    current_db_dataframe.to_pickle('using crm crafted method.pkl')
```

```python
#this helps to generate a results.csv which tells names and their potential duplicates using fuzzy/proximity search 
pros(df[["full"]])  # this produces an output where in the last column, all the names that are similar to the name in the first column are displayed
```

## Method 2 using all the methods given in fuzzy wuzzy explicitly

# Using token_set_ratio approach. Here, this approach has been chosen out of ratio, partial ratio and token_sort_ratio because comparison is done as follows:

t0 = [SORTED_INTERSECTION]<br>
t1 = [SORTED_INTERSECTION]+[SORTED_REST_OF_STRING1]<br>
t2 = [SORTED_INTERSECTION]+[SORTED_REST_OF_STRING2]

### The intuition here is that because the SORTED_INTERSECTION component is always exactly the same, the scores increase when (a) that makes up a larger percentage of the full string, and (b) the string remainders are more similar. This takes into account abbreviations, letters added at the end & letters added in between the names which change the spelling of the names but imply the same person. The similarity between full names should be >=90. Each entry in the dataframe is compared with every other entry after it.

#### References: 
https://marcobonzanini.com/2015/02/25/fuzzy-string-matching-in-python/

http://chairnerd.seatgeek.com/fuzzywuzzy-fuzzy-string-matching-in-python/

#### Here, for 1st approach, it is considered that different dates of birth and different genders mean different patients even for same full name. We iterate over each patient and find out how much similar the full names of the patients coming after him in the dataframe are provided gender and dob are same.

```python
#assuming that different dates of birth and different genders mean different patients even for same full name
def tokenset(df):
    tot=len(df)
    for i,r in df.iterrows():
        dobm=r["dob"]
        gm=r["gn"]
        fnm=r["full"]
        if(r["keep"]!=False):
            for j in range(i+1,tot):
                rc=df.iloc[j]
                dobc=rc["dob"]
                gc=rc["gn"]
                fnc=rc["full"]
                if(dobm==dobc and gm==gc):
                    if(fuzz.token_set_ratio(fnm,fnc)>90):
                        df.set_value(j,"keep",False)
    df=df[df["keep"]==True]
    df=df.reset_index(drop=True)
    print(df)
    print(df.shape)
    df.to_pickle("using token sort by unique dob and gender.pkl")
    df=df.drop(["full","keep"],axis=1)
    df.to_csv("using token sort by unique dob and gender.csv", index=False)
    #we dropped from 62 rows to 51 rows
tokenset(df)
```
